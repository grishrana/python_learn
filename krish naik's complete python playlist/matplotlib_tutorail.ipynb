## **Matplotlib Tutorial**

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

### **Scatter plot**

```python
plt.scatter(x,y,c='g') # c stands for color
plt.xlabel('X axis')
plt.ylabel("Y axis")
plt.title("Graph in 2D")
plt.savefig("Test.png")
```

```python
plt.plot(x,x*x,'ro:')
```

```python
plt.plot(x,y,'r',linestyle='dashed',markersize=12,linewidth=2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Another 2D diagram")
```

```python
plt.subplot(2,2,1)
plt.plot(x,x,'r')

plt.subplot(2,2,2)
plt.plot(y,y,'g')

plt.subplot(2,2,3)
plt.plot(x,x*x,'b')

plt.subplot(2,2,4)
plt.plot(y,y**y,':r')
```

```python
x=np.arange(1,11)
y= 3*x+5
plt.title("Matplotlib Demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
```

```python
np.pi
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
plt.title("Sine wave graph")
plt.plot(x,y)

```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.title("Sine wave")
plt.plot(x,y_sin,"r:")

plt.subplot(2,1,2)
plt.title("Cos wave")
plt.plot(x,y_cos,"g--")
```

### **Barplot**

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

### **Histogram**

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a)
plt.title('Histogram')
```

### **Boxplot**

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
```

```python
data
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

### **Pie Chart**

```python
labels= 'Python','C++','Ruby', 'Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0.2,0) #explode slices

#plot
plt.pie(sizes,explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False)

plt.axis('equal')
```
