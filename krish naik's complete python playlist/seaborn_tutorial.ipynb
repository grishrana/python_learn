## **Seaborn Tutorial**

```python
import seaborn as sns
import numpy as np
```

### **Correlation with Heatmap**

A correlation heatmap uses colored cells, typically in a monochormatic scale, to show a 2D correlation matrix(table) between two discrete or event types. It is very important in Feature Selection.

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

## **Jointpot**
A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

### **Univariate Analysis**

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## **Pair plot**

A `pair plot` is also known as scatterplot, in which one variable is the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.pairplot(df,hue='sex')
```

## **Dist Plot**

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# **Categorical Plots**
1. Countplot
2. Violinplot
3. barplot


```python

sns.countplot(df['sex'])
```

```python
sns.countplot(x='smoker',data=df,hue='smoker')
```

```python
sns.countplot(x='day',data=df,hue='day')
```

## **Barplot**

```python
sns.barplot(x='total_bill',y='sex',data=df,hue='sex')
```

```python
sns.barplot(x='smoker',y='total_bill',data=df,hue='smoker')
```

### **Box plot**

```python
sns.boxplot(data=df,x='smoker',y='total_bill',palette='rainbow')
```

```python
sns.boxplot(data=df,x='sex',y='tip',hue='sex')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## **Violinplot**
Violin plot helps us to see both the distribution of data in terms of kernel desnity estimation and the box plot.

```python
sns.violinplot(x="total_bill",y="day", data=df, palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=df)
```
