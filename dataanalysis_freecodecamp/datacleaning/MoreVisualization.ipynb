# **More Visualization**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

There are two API for matplotlib: `Global` and `OOP`. Which have thier own pros and cons.

### **Global Api**
Matplotlib's default pyplot API has a global.

```python
x=np.arange(-9,10)
```

```python
plt.figure(figsize=(12,6))
plt.title('My nice plot')
plt.plot(x,x**2)
plt.plot(x,-1*(x**2))
plt.legend(['x**2','-(x**2)'])
```

```python
plt.figure(figsize=(12,6))
plt.title('My Nice plot')

#1 row and 2 columns are created and 1st panel is selected
plt.subplot(1,2,1) #no of rows, no 0f columns, selected panel
plt.plot(x,x**2)
plt.plot([-9,0,9],[0,0,0]) #ploting a straight line
plt.legend(['X^2','Horixontal line'])
plt.xlabel('X')
plt.ylabel('X squared')

#for 2nd diagram lets choose the 2nd panel
plt.subplot(1,2,2)
plt.plot(x,-1*(x**2))
plt.plot([0,0,0],[-81,-40,0])
plt.legend(['-X^2','Vertical line'])
plt.xlabel('X')
plt.ylabel('-X squared')
```

### **OOP Interface**

```python
fig, axes =plt.subplots(figsize=(12,6))
axes.plot(
    x,x**2,color='red',linewidth=3,
    marker='o', markersize=6,label='x^2'
          )

axes.plot(
    x,-(x**2),'b-.',label='-x^2',linewidth=4
)

axes.set_xlabel('x')
axes.set_ylabel('X squared')
axes.set_title('My nice plot')
axes.legend()
fig
```

```python
fig,axes=plt.subplots(figsize=(12,6))
axes.plot(x,x+0,linestyle='solid',label='x=x') #linestyle='solid can be replaced by '-'
#i.e axes.plot(x,x+0,'-')

axes.plot(x,x+1,linestyle='dashed',label='x=x+1') #likewise linestyle='dashed' can be replaced by '--'
axes.plot(x,x+2,linestyle='dashdot', label='x=x+2') # '-.'
axes.plot(x,x+3,linestyle='dotted', label='x=x+3') # ':'
axes.legend()
axes.set_title('My nice Plot')

fig
```

#### **Format Strings**

A format string consists of a part of color, marker and line:

`fmt = '[marker][line][color]'` or `fmt='[color][marker][line]'`
Each of them is optional. If not provided, the value form the style cycle is used. Exception: If `line` is given, but no `marker`, the data will be a line without markers.


We can reduce the code <br>
__*axes.plot( <br>
    x,x**2,color='red',linewidth=3, marker='o', markersize=6, label='x^2' <br>
          )*__ <br>
as <br>
*__axes.plot( <br>
    x, x**2, linewidth=3, ':r', markersize=6, label='x^2' <br>
          )__* <br>
here `:r` represents red __dotted lines__. We can combine the `linestyle` parameter and `color` parameter into a single argument.

```python
fig, axes=plt.subplots(figsize=(12,6))
axes.plot(x,x+0,'-g',label='solid green line')
axes.plot(x,x+1,':c',label='dotted cyan line')
axes.plot(x,x+2,'o-r',label='dotted solid red line') # {o: circle marker, -: solid line style, r: red}
axes.plot(x,x+3,'-.b',label='dotted dash blue line')
axes.legend()
axes.set_title('My Nice Plot')
```

Other various combination of `'[marker][line][color]'` can be used.

```python
print(f"Markers: {[m for m in plt.Line2D.markers]}")
```

```python
print(f"Line:{[m for m in plt.Line2D.lineStyles]}")
```

### **Other Types of plots**

#### **Figures and Subfigures**
When we call the `subplots()` function we get a tuple containing a `Figure` and a `axes` element.

```python
plot_objects=plt.subplots()
figure, ax = plot_objects
ax.plot([1,2,3],[0,5,6])
print(plot_objects)
```

We can also define how many elements we want inside our figure. To do that we can set the `nrows` and `ncols` parameters:

```python
plot_objects=plt.subplots(nrows=2,ncols=2,figsize=(12,6))
fig, ((ax1,ax2),(ax3,ax4))=plot_objects
plot_objects
```

```python
ax1.plot(np.random.randn(20),'--b')
ax4.plot(np.random.randn(20),':y')
ax2.plot(np.random.randn(30),'o-b',linewidth=3.0)
ax3.plot(np.random.randn(20),',-.c',linewidth=4.0)

fig
```

#### **The `subplot2grid()` function**

Create a subplot at a specific location inside a regular grid.

```python
fig=plt.figure(figsize=(15,8))
ax1=plt.subplot2grid((3,3),(0,0),colspan=3)
ax2=plt.subplot2grid((3,3),(1,0),colspan=2)
ax3=plt.subplot2grid((3,3),(1,2),rowspan=2)
ax4=plt.subplot2grid((3,3),(2,0))
ax5=plt.subplot2grid((3,3),(2,1))
```

#### **Scatter Plot**

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=np.pi*(20*np.random.rand(N))**2
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x,y,s=area,c=colors,alpha=0.5,cmap='viridis')
plt.colorbar()

plt.show()
```

```python
fig=plt.figure(figsize=(12,6))
ax1=fig.add_subplot(1,2,1)
plt.scatter(x,y,s=area,c=colors,alpha=0.5,cmap='Spectral')
plt.colorbar()

ax2=fig.add_subplot(1,2,2)
plt.scatter(y,x,s=area,c=colors,alpha=0.6,cmap='Pastel1')
plt.colorbar()

plt.show()
```

#### **Histograms**

```python
Value=np.random.randn(300)
print(Value)
```

```python
plt.subplots(figsize=(12,6))
plt.hist(Value,bins=100,alpha=0.8,histtype='bar',color='olive',edgecolor='black')
#bins=100 means the np array Value is divided into 100 parts and plotted.
plt.xlim(xmin=-5,xmax=5)

plt.show
```

#### **KDE(Kernal Density Estimation)**

```python
from scipy import stats

density=stats.gaussian_kde(Value)
density
```

```python
plt.subplots(figsize=(12,6))

value2=np.linspace(min(Value)-10,max(Value)+10,100)

plt.plot(value2,density(value2),color='green')
plt.fill_between(value2,0,density(value2),alpha=0.3,color='olive')
plt.xlim(xmin=-5,xmax=5)
plt.show()
```

#### **Combine plots**

```python
plt.subplots(figsize=(12,6))

plt.hist(Value,bins=100,color='olive',alpha=0.8,edgecolor='black',density=1,histtype='bar')

plt.plot(value2,density(value2),color='green',linewidth=4.0)
plt.xlim(xmin=-5,xmax=5)

plt.show()
```

#### **Bar Plots**

```python
Y=np.random.rand(1,5)[0]
Y2=np.random.rand(1,5)[0]
print(Y,Y2)
```

```python
plt.figure(figsize=(12,6))

plt.bar(np.arange(len(Y)),Y,width=0.5,color='Violet')
plt.show
```

```python
plt.figure(figsize=(12,4))

plt.bar(np.arange(len(Y)),Y,width=0.5,color='Violet', label='Label Y',edgecolor='green')
plt.bar(np.arange(len(Y2)),Y2,width=0.5,color='Cyan',bottom=Y,label="Label Y2")

plt.legend()
plt.show()
```
